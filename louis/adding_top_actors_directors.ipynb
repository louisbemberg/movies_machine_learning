```python
import numpy as np
import pandas as pd
```

```python
# load original csv containing information about directors/actors
crew_names = pd.read_csv("name_basics.csv", sep='\t')
crew_names
```

```python
# Loading names of top 100 actors and top 200 directors from IMDB
top_actors = pd.read_csv("best_actors.csv")
top_directors = pd.read_csv("best_directors.csv")
```

```python
# removing surrounding whitespace and turning everything into lowercase
crew_names['primaryName'] = crew_names['primaryName'].apply(str.strip).apply(str.lower)
top_directors["# director_name"] = top_directors["# director_name"].apply(str.strip).apply(str.lower)
top_actors["# actor_name"] = top_actors["# actor_name"].apply(str.strip).apply(str.lower)
```

```python
# turning top actors into python lists for comparison in next cell
top_director_names = list(top_directors["# director_name"])
top_actor_names = list(top_actors["# actor_name"])
```

```python
# creating binary columns "is_top", checking whether the director/actor is in the top list
crew_names['is_top_actor'] = np.where(crew_names['primaryName'].isin(top_actor_names), 1, 0)
crew_names['is_top_director'] = np.where(crew_names['primaryName'].isin(top_director_names), 1, 0)
```

```python
# how many top actors are there in the df?
crew_names['is_top_actor'].value_counts()
```

```python
# how many top directors are there in the df?
crew_names['is_top_director'].value_counts()
```

```python
# exporting new column to CSV
crew_names.to_csv('name_basics_improved.csv', encoding='utf-8', index=False)
```
