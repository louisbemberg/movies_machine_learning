```python
import numpy as np
import pandas as pd
import csv
```

```python
title_akas = pd.read_csv("title_akas.tsv", sep='\t')
```

```python
us_movies=title_akas[title_akas['region']=='US']
```

```python
us_movies.groupby('types').sum()
```

```python
us_movies.groupby('language').sum()
```

```python
grouped = us_movies.groupby('attributes').sum()
grouped.sort_values('ordering', ascending=False)
```

```python
usMoviesID=us_movies['titleId']
```

```python
title_basics = pd.read_csv("title_basics.tsv", sep='\t')
```

```python
titleBasicsUS = title_basics[title_basics['tconst'].isin(usMoviesID)]
```

```python
longUsMovies = titleBasicsUS[titleBasicsUS['titleType']=='movie']
df = longUsMovies['startYear'].apply(pd.to_numeric, errors='coerce')
longUsMovies = longUsMovies.assign(startYear = df)
LongUsMoviesWoNa = longUsMovies.dropna(axis = 0, subset = ['startYear'])
```

```python
movieSelection = LongUsMoviesWoNa[LongUsMoviesWoNa['startYear'] > 1990]
len(movieSelection)
```

```python
movieSelection.groupby('endYear').count() 
```

```python
movies_selection_filtered = movieSelection.drop(columns=["endYear"])
```

```python
movies_selection_filtered.to_csv("movies_selection.csv", index=False)
movies_selection_filtered
```

```python
read_movies_selection = pd.read_csv("movies_selection.csv")
read_movies_selection
```

```python
read_movies_selection['genres'] = read_movies_selection['genres'].apply(lambda x: x.split(','))
```

```python
read_movies_selection
```

```python
#read_movies_selection_dummies = pd.get_dummies(read_movies_selection, columns=['genres'], prefix=['is'])

read_movies_selection_dummies = pd.get_dummies(read_movies_selection.genres.apply(pd.Series).stack()).sum(level=0)
```

```python
read_movies_selection_dummies.sum() < 5000
```

```python
filtered = read_movies_selection_dummies.loc[:, read_movies_selection_dummies.sum() < 5000]
filtered = filtered.drop(columns=[r'\N'])
filtered
```

```python
read_movies_selection
```

```python
read_movies_selection = read_movies_selection.join(filtered)
read_movies_selection = read_movies_selection.drop(columns=['genres', 'titleType'])
read_movies_selection
```

```python
read_movies_selection.to_csv("movies_selection_categorical.csv", index=False)
```
