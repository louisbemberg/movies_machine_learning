## Final Data

```python
import pandas as pd
```

```python
df = pd.read_csv("../00_DATA/final_filtered_data.csv")
df = df.drop(columns=["Unnamed: 0", "Unnamed: 0.1"], errors="ignore")
df
```

### Years since produced (reference: 2021)

```python
df["yearsSinceProduced"] = 2021 - df["startYear"]
df.groupby("yearsSinceProduced").count()
```

## Data Splitting

```python
df_train = pd.DataFrame(df[df["startYear"] <= 2015])
df_test = pd.DataFrame(df[df["startYear"] > 2015])
assert(df_train.shape[0] + df_test.shape[0] == df.shape[0])
```

```python
df_train["isTopActor"] = pd.Series([1 if x == True else 0 for x in df_train["isTopActor"]], index=df_train["isTopActor"].index)
df_train["isTopDirector"] = pd.Series([1 if x == True else 0 for x in df_train["isTopDirector"]], index=df_train["isTopDirector"].index)

df_test["isTopActor"] = pd.Series([1 if x == True else 0 for x in df_test["isTopActor"]], index=df_test["isTopActor"].index)
df_test["isTopDirector"] = pd.Series([1 if x == True else 0 for x in df_test["isTopDirector"]], index=df_test["isTopDirector"].index)
df_test
```

```python
# Rename columns
df_train = df_train.rename(columns={'Sci-Fi': 'SciFi', 'Talk-Show': 'TalkShow', 'Reality-TV': 'RealityTV', 'Game-Show': 'GameShow'})
df_test = df_test.rename(columns={'Sci-Fi': 'SciFi', 'Talk-Show': 'TalkShow', 'Reality-TV': 'RealityTV', 'Game-Show': 'GameShow'})
```

```python
df_train = df_train.drop(columns=["Adult", "isAdult", "RealityTV", "Short", "TalkShow"])
df_test = df_test.drop(columns=["Adult", "isAdult", "RealityTV", "Short", "TalkShow"])
df_train
```

```python
df_train.to_csv("train_dataset.csv", index=False)
df_test.to_csv("test_dataset.csv", index=False)
```
